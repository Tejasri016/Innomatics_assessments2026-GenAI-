1. User Login Check

```python
username = "admin"
password = "1234"
if username == "admin" and password == "1234":
    ##if the username and password both are matches the following statement prints
    print("Login successful")
else:##if the above if condition not satisfies then the below statement will prints
    print("Invalid credentials")
```

2. Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
passcount = 0 ##initially passcount value is 0
failcount = 0 ##initially failcount value is 0
for i in marks:
    if i >= 50:
        ##if the marks of the student greaterthan or equal to 50 , passcount value increase '+1'
        passcount +=1
    else:
        ##if the marks of the student doesn't satisfies the if condition ,failcount value increase '+1'
        failcount +=1
print(f"Total no of pass students are :{passcount}")
print(f"Total no of fail students are :{failcount}")
```

3. Simple Data Cleaner

```python
names = ["Alice ", "bob", "CHARLIE"]
lowercase = []
for n in names:
    lowercase.append(n.strip().lower())
    ##lower converts the strings into lowercase
    ##strip removes the unwanted whitespaces in the names
    ##append -> we use this to add elements into the array 
print(lowercase)
```

4. Message Length analyzer

```python
message = ["Hi", "Welcome to the platform", "OK"]
for msg in message:
    length = len(msg) ##here this line calculates the characters of the msg 
    print(f"{msg} --> this msg length is : {length}")
    flag = len(msg) > 10
    ##if the msg contains more than 10 characters then it returns the flag value as 'True' else 'False'
    print(f"flag : {flag}")
```

5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count = 0 ##initially count value is 0
for entries in logs:
    if entries == "ERROR":
        ##if the entries == 'ERROR' then the count value increases '+1'
        count +=1
print(f"The no of (ERROR) entries in the given logs are : {count}")
        
```
