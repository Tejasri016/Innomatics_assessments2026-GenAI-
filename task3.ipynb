Problem 1: Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi" : 92,
    "Anita" : 88,
    "Kiran" : 92,
    "Suresh" : 85
}
#finding maximum score of employees
max_score = max(employees.values())
#finding employees with maximum score by iterating through the dictionary 
top_performers = [name for name in employees if employees[name]==max_score]
# printing the top performers and their score 
# for better appearance we join the list of top performers with comma & space
print("Top Performers eligible for Bonus: {}".format(",".join(top_performers)),"(score:{max_score})")
```

Problem 2: Search Query Keyword Analysis

```python
products = ["Phone", "Laptop", "Tablet", "  Phone", "Laptop", "Headset ", "Tablet"]
product_counts = {}
#iterating through the products list
for product in products:
    # Removing duplicates and striping whitespace from products list using strip() & set() functions
    product = product.strip().lower()
    # counting the occurance of each product in the list and storing it in a dictionary
    product_counts [product] = product_counts.get(product, 0) + 1
#printing unique products with its count 
print(product_counts)
```

Problem 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
# using list comprehsion to filter out the gmail addresses from the list of emails
gmail_addresses = [email for email in emails if email.endswith("@gmail.com")]
# filtering out the yahoo addresses from the list of emails 
yahoo_addresses = [email for email in emails if email.endswith("@yahoo.com")]
# calculating the percentages of both gmail & yahoo addresses
gmail_percent = (len(gmail_addresses)/len(emails))*100
yahoo_percent = (len(yahoo_addresses)/len(emails))*100
print(f"Gmail addresses: {gmail_percent}%")
print(f"Yahoo addresses: {yahoo_percent}%")
```

Problem 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
# claculating the average sales
average_sales = sum(sales)/len(sales)
#print(average_sales)
# identifying the days where sales were above 30% of the average_sales 
#we use enemurate() function to get the index of the day and value of the sale 
for i, sale in enumerate(sales, start=1):
    if sale > average_sales:
        print(f"Day {i}: {sale}")
```

Problem 6: Duplicate User ID Detection  

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
# here we created a dictionary to count the occurance of each user id in the list
count_user_ids = {}
# here we created a dictionary to store the duplicates user ids and their count
duplicates = {}
# for loop to iterate through the user_ids list
for user in user_ids:
    # counting the occurances of each user id
    count_user_ids[user] = count_user_ids.get(user, 0)+1
    # if the count of user is > 1 the user is duplicate and we store it in the duplicates dictionary with its count
    if count_user_ids[user] >1 :
        duplicates[user]=count_user_ids[user]
# printing the duplicate user ids with its repeated count for that we use items() fucntion to get the key and value of the user id & its count from the duplicates dictionary
for user, count in duplicates.items():
    print(f"{user} -> {count} times")
```
