problem 1: Unique words in a sentence

```python
sentence = "python is easy and python is powerful"
s = sentence.split()# here split function is used to split the sentence into words and store it in a list
print(s)
# ['python', 'is', 'easy', 'and', 'python', 'is', 'powerful']
sen = set(s) # here set function is used to remove the duplicate words from the list and store it in a set
print(f"Unique words count:{len(sen)}") # here len function is used to count the number of unique words present in the set
print(f"Unique Words : {sen}") #here we are printing the unique word set
```

problem 2: highest salary from Employee Data

```python
employees = {
    "Ravi" : 75000,
    "Anita" : 68000,
    "Kiran" : 72000
}
highest_salary_employee = max(employees) # here max function is used to find the maximum salary employee from the dictionary and return the key of the maximum salary employee
salary = max(employees.values()) # here max function is used to find the maximum salary from the dictionary and return the key of the maximum salary 
print(f"Highest Salary: {highest_salary_employee} - {salary}")
```

problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]
max_val = max(numbers) # here max function is used to find the maximum value in the list
min_val = min(numbers) #here min function is used to find the min value of the list 
print(f"List : {numbers}")
print(f"Max : {max_val}")
print(f"Min : {min_val}")
```

Problem 4: Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]
count=0 #initially count is set to 0 and then we are iterating through the list of prices 
for p in prices:
    if p > 1000: # here we are checking if the price is greater than 1000 then we are incrementing the count value by 1
        count+=1
print(f"Products above 1000 : {count}")
```
