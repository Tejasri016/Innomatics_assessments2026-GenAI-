```python
#Problem 1: Smart Parking Lot Management System

def smart_parking_system(vehicles):
    parking_capacity=50
    current_parked_vehicles = 0
    peak_usage = 0
    # iterating each vehicle in vehicles list
    for vehicle in vehicles:
        # check the vehicle parked or exited 
        if vehicle == "IN":
            current_parked_vehicles+=1
        elif vehicle == "OUT":
            if current_parked_vehicles > 0:
                current_parked_vehicles-=1 
        
     # peak usage checking
        if current_parked_vehicles > peak_usage:
            peak_usage = current_parked_vehicles 
    print(f"Currently parked vehicles:{current_parked_vehicles}")
    print(f"Peak parking usage: {peak_usage}")   
    # Alert condition   
    if current_parked_vehicles > parking_capacity:
        print("Parking Status: Not Available")
    else:
        print("Parking Status: Available")
# input data   
vehicles = ["IN", "OUT", "IN", "IN", "OUT"]
# call the function
smart_parking_system(vehicles)

     


```

```python
#Problem 2: Online Food Delivery Time Estimator

def online_delivery_time_detector():
    distance = int(input("enter the distance:"))
    weather_condition = input("choose weather condition(rainy/sunny/normal):").lower()
    traffic_level = input("choose traffic level(high/medium/low):").lower()
    # check distance
    if distance <= 0:
        print("Invalid distance")
        return
    # traffic level
    if traffic_level == "low":
        multiplier = 5
    elif traffic_level == "medium":
        multiplier = 8
    elif traffic_level == "high":
        multiplier = 14
    else:
        print("invalid traffic level")

    # check weather condition
    if weather_condition == "rainy":
        multiplier+=5
    elif weather_condition == "sunny":
        multiplier += 3
    elif weather_condition == "normal":
        multiplier += 0
    else:
        print("Invalid weather condition")
    # calculating delivery time
    delivery = distance * multiplier
    hours = delivery//60
    minutes = delivery%60
    print(f"Estimated delivery time: {hours}hrs {minutes}mins")
# function calling
online_delivery_time_detector()
```

```python
#Problem 3: Movie Theatre Seat Occupancy Analyzer

# create a function to check the movie seats availabiltiy of theatre
def movie_seat_occupancy():
    total_seats = 200
    booked_seats = [1]*150
    # length of booked seats
    seat_count = len(booked_seats)
    # calculate percentage
    percentage = (seat_count/total_seats)*100
    # check status using percentage & suggests opening additional shows
    if percentage == 100:
        status = "House full"
        suggestion = "open additional show immediately"
    elif percentage >=75 :
        status = "Almost full"
        suggestion = "consider additional show"
    else:
        status = "Seats are available, you can book now..!"
        suggestion = "No need for additional show"
    return percentage,status,suggestion
#function calling and printing the reaults
percentage,status,suggestion = movie_seat_occupancy()
print(f"Occupancy: {percentage}%")
print("show Status: ",status)
print("Suggestion: ",suggestion)
        
```

```python
#Problem 5: Smart Classroom Resource Usage Monitor

def resource_usage_monitor(resources):
    limit_hrs = 7
    for resource,value in resources.items():
        if value >= limit_hrs:
            status = "yes"
        else:
            status = "no"
        return resource,status
#input data
resources = {
    "Ac" : 9,
    "Projector" : 6,
    "Lights" : 4
}
# function calling
resource,status = resource_usage_monitor(resources)
print("Overused Resources:",resource)
print("Energy Alert: ",status)

```

```python
# problem 6: Online Event Registration Capacity Controller

#create a fucntion that manages event registartions
def registration_capacity_controller():
    # give event capacity
    event_capacity = 100
    # give no of registrations
    registrations = 105
    # initialize waitlist & confirmed as '0'
    waitlist = 0
    confirmed = 0
    # check registrations with capacity
    #  then return how many registrations will be in waitlist and confirmed
    if registrations >= event_capacity:
        waitlist = registrations - event_capacity
        confirmed = registrations - waitlist
        status = "closed"
    else:
        waitlist = 0
        confirmed = registrations
        status = "Open"
    return waitlist,confirmed,status
# function calling
waitlist,confirmed,status = registration_capacity_controller()
print(f"Confirmed Registrations: {confirmed}")
print(f"Waited users:{waitlist}")
print(f"Registration Status: {status}")
```
