Problem 1: Mobile Recharge Validation System

```python
# recharge plan validation
mobile_recharge_plan = int(input("Enter the recharge plan amount(199, 299, 399, 599):"))
# here while condition checks that if the plan amount is >= 50 then it will check the plan is valid or not
while mobile_recharge_plan >= 50:
    # by using if-elif-else statement we can check the plan validation
    if mobile_recharge_plan == 199:
        print("You have selected the 199 recharge plan")
        break
    elif mobile_recharge_plan == 299:
        print("You have selected the 299 recharge plan")
        break
    elif mobile_recharge_plan == 399:
        print("You have selected the 399 recharge plan")
        break
    elif mobile_recharge_plan ==  599:
        print("You have selected the 599 recharge plan")
        break
    # if the plan amount is not vallid then it will print the amount of plan is invalid and then we should retry to echange the plan & then make it valid
    else:
        print("Invalid recharge plan amount : Please retry the rechange the plan amount")
        mobile_recharge_plan = int(input("Enter the recharge plan amount(199, 299, 399, 599):"))
    
```

Problem 2: Inventory Reorder Alert System

```python
# stock alert for products
def stock_alert(products):
    # this function checks the stock levels of products and prints a statement that indicating whether we can reorder or not
    for product, stock in products.items():
        # if the stock level less than 15, we should reorder the product
        if stock < 15:
            print(f"Reorder alert for {product}")
        
        # otherwise the stock level is ok that means we don't need reorder the product
        else:
            print(f"Stock ok for {product}")
# example usage of stock_alert function
# here we ahve a dictioanary of products and their stock quantities
products = {
    "product A " : 10,
    "Product B" : 20,
    "Product C " : 15,
 }
stock_alert(products)
```

Problem 3: Student Result Processing

```python
# student result validation
def marks_validation(student_marks):
    # this function calculated the average marks through the for loop 
    for marks in student_marks:
        average = sum(student_marks) / len(student_marks)
    # if the average amrks of a student is >= 50 then it returns that student passed
    if average >= 50:
        print("Student passed")
    # otherwise the student failed
    else:
        print("Student failed")
# here we have a list of student marks and we will pass this list to the marks_validation function to check the final result of the student
student_marks = [69, 78, 89, 90, 87]
marks_validation(student_marks)
```

Problem 4: Cab Fare Estimator with Retry Option

```python
# creating a cab fare calculator
def cab_fare_calculator(distance, is_peak=False):
    base_fare = 50
    per_km = 12
    # calculating the total fare by using formula
    total_fare = base_fare + (per_km * distance)
    # if peak hour extra 25% added to the total_fare
    if is_peak:
        total_fare *= 1.25
    return total_fare
while True:
    try:
        distance = float(input("enter the distance of the ride:"))
        is_peak_input = input("Is it peak hour? (yes/No):").lower()
        is_peak = is_peak_input == "yes"
        fare = cab_fare_calculator(distance, is_peak)
        print(f"The total cab fare is: {fare}")
        break
    except ValueError:
        print("Invalid input: please enter a valid number for distance and yes or no for peak hour")
        
```

Problem 5: employee Attendeance Eligibility Checker

```python
# create a function to check the employee attendance
def employee_attendance_checker(employee_attendance):
    # initially count value is 0
    count = 0
    # iterating the employees attendance by using for loop
    for attendance in employee_attendance:
        # if the value is 'P' then count vlaue increment by 1
        if attendance == 'P':
            count += 1
        # calculating attendance percentage
    res = (count / len(employee_attendance))*100
    # check if the attendance percentage is greater thyan 75 or not
    if res >= 75:
        return "Employee is eligible for promotion"
    else:
        return "Employee is not eligible"
#here list of employee attendance
employee_attendance = [ 'P', 'A' , 'P', 'P', 'A']
print(employee_attendance_checker(employee_attendance))
```

Problem 6: Password Strength Checker

```python
def password_checker(password):
    min_len = 8
    has_special = False
    has_digit = False
    specialchars = "@#$"
    # check minimum length
    if len(password)< min_len:
        return "Password must contain min_length of 8"
    # iterating the password for each character
    for p in password:
        # if the char is a digit then it returns true
        if p.isdigit():
            has_digit = True
        # if the char is in specialcharacters then it return true
        if p in specialchars:
            has_special = True
    # if the password not contain any digit return the statement that the password must contain atleast 1digit
    if not has_digit:
        return False, "password must contain atleast one digit"
    # if the password not contain any special characters then return the statement that the password must contain atleast one special symbol
    if not has_special:
        return False, "password must contain atleast one special character"
    # whatever the rules for the password like min_len 8 , atleast 1 special char & 1 digit then return the below statement 
    return "password is valid"

# taking input out of the function
password = input("enter your password:")
check = password_checker(password)
print(check)

        
        
```
