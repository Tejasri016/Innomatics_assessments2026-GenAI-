Problem 1: Social Media - Post Engagement Analysis

```python
def post_engagement_analysis(likes):
    total = 0
    for like in likes.values():
        total += like
    print(f"Total likes: {total}")
    if total >= 1000:
        return "Post status: Viral Post"
    else:
        return "Post Status: Normal Engagement"
    
likes = {
    "p1":250,
    "p2":300,
    "p3":400,
    "p4":500
}
result = post_engagement_analysis(likes)
print(result)
```

Problem 3: Agriculture - Rainfall Adequacy Checker

```python
# create function for rainfall analysis
def rainfall_adequacy_checker(seasonal_rainfall):
    # required level of rainfall
    required_level = 55
    # calculate average of rainfall for all seasons
    total_rainfall = sum(seasonal_rainfall.values())
    no_of_seasons = len(seasonal_rainfall)
    average = total_rainfall/no_of_seasons
    #print average
    print(f"Average Rainfall :{average}")
    # check if the average rainfall greater than required_level or not
    if average >= required_level:
        return "Rainfall Status: Adequate Rainfall"
    else:
        return "Rainfall Status: Inadequate Rainfall"
# rainfall seasonal data
seasonal_rainfall={
    "rainy": 100,
    "winter" : 50,
    "Summer" : 60
}
# call the function
result = rainfall_adequacy_checker(seasonal_rainfall)
print(result)
```

Problem 4: Social Media - Duplicate Account Detection

```python
#create a function for duplicate account detection
def duplicate_account_detector(account):
    # create empty set for duplicates
    duplicates = {}
    # iterating each account
    for duplicate in account:
        duplicates[duplicate] = duplicates.get(duplicate, 0)+1
    # check if account have more than one or not by its count
    for count in duplicates.values():
        if count > 1:
            return "Duplicate Accounts Found: Yes"
        else:
            return "Duplicate Accounts Found: No"
    return count
# input of accounts
account=["aruncode_12", "ishan_creation","aruncode_12"]
# call the function
result = duplicate_account_detector(account)
print(result)
```

Problem 7: System Monitoring - Application Health Checker

```python
# create fucntion to count health based errors
def application_health_checker(health):
    # count is set to be as '0' initially
    count = 0
    # iterating each status in health
    for status in health:
        # check if status error or not if error found then increment count
        if status == "Error":
            count +=1
    print(f"Error count:{count}")
    # check health status by count of errors
    if count == 0 :
        print("System status: Healthy")
    elif count <= 5:
        print("System status: Minor issues")
    else:
        print("System status : Critical issues")
# here list of health status
health = ['Error','NoError','Error','Error','NoError']
# call the function
application_health_checker(health)
```

Problem 8: Banking - Daily Transaction Limit Checker

```python
# create a function for checking daily transactions
def daily_transaction_checker():
    # initia;ize total as '0'
    total = 0
    # initialize daily limit
    daily_limit = 50000
    # ask how many transactions
    n = int(input("enter no of transactions you did: "))
    # iterating each transaction throughh for loop
    for i in range(n):
        # ask amount
        daily_expense = int(input(f"enter the amount (T{i}):"))
        # sum up the transaction amount
        total += daily_expense
    print(f"Transaction amount: {total}")
    # check if the transactions within dailylimit or not
    if total <= daily_limit:
        return "Transaction Status: Approved"
    else:
        return "Transaction status: Rejected"
# call the function
result = daily_transaction_checker()
print(result)


```

Problem 9: E-Learning - Student Attendance Eligibility

```python
# create a function to check student attendance percentage & check exam eligibilty
def check_exam_eligibility():
    # ask no of days present & absent
    present = int(input("enter no.of present days:"))
    absent = int(input("enter no.of days absent:"))
    #calculate total of attendance
    total = present+absent
    # calculate percenatge
    percentage = (present/total)*100
    print(f"Attendance percentage: {percentage}")
    # check if the percentage >= 75% or not
    if percentage >= 75:
        return "Exam Eligibilty: Eligible"
    else:
        return "Exam Eligibility: Not Eligible"
#call the function
result = check_exam_eligibility()
print(result)
```

Problem 10: Smart Electricity Bill Analyzer

```python
# create function for calculating electricity bill
def electricity_bill_analyzer(units):
    # apply billing rules using conditions
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100*3) + (units-100)*5
    else:
        bill = (100*3) + (100*5) + (units - 200)*7
    return bill
# input of units consumed
units = int(input("enter the consumed units :"))
# call the function to print total bill
total_bill = electricity_bill_analyzer(units)
print(f"Total electricity bill: {total_bill}")
# check usage status 
if total_bill < 500:
    print("Usage status: Low Usage")
elif total_bill >= 500 and total_bill <= 1500:
    print("Usage status: Moderate Usage")
else:
    print("Usage status: High Usage")


        
```
