```python
#Problem 2: Employee ID Card System

# create class to store empoloyee details
class EmployeeId:
    # function for name,id, department
    def __init__(self,name,id,dept):
        self.name = name
        self.id = id
        self.dept = dept
        # display results
    def show_result(self):
        print("Employee Id Card")
        print("Name: ",self.name)
        print("ID:",self.id)
        print("Department:",self.dept)
#object creation
employee = EmployeeId("Rahul","EMP123","IT")
# call the funtion 
employee.show_result()
```

```python
# problem 3: Mobile contact record system

# create class to store mobile contact record
class MobileContact:
    def __init__(self,name,phno):
        self.name = name
        self.phno= phno
    def status(self):
        print("Contact Saved")
        print("Name:",self.name)
        print("Phone number:",self.phno)
#object
contact = MobileContact("Isha",6356789201)
contact.status()
```

```python
#Problem 4: Product Price Tag Generator

# create class to store product details
class Product:
    # fucntuion for product name & price
    def __init__(self,name,price):
        self.name=name
        self.price=price
        # display results
    def result(self):
        print("Product:",self.name)
        print(f"Price:{self.price}/-")
#object
productname=Product("Headphones",2500)
# call the function
productname.result()
```

```python
# Problem 5: Movie Rating Display System

# craete a class to store movie ratings
class MovieRating:
    # fucntion to store movie name & rating
    def __init__(self,movie,rate):
        self.movie=movie
        self.rate=rate
        # display results
    def result(self):
        print("Movie:",self.movie)
        print(f"Rating: {self.rate} /5")
#object
movierating = MovieRating("Sarvam Maya",4.9)
# call the function
movierating.result()
```
